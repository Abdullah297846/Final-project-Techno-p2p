```python
df_convert = pd.get_dummies(loan_df['LoanStatus'])
print(loan_df.head())
```

```python
df_af = pd.concat((df_convert, loan_df), axis=1)
df_af = df_af.drop(["LoanStatus"], axis=1)
df_af = df_af.drop(["Cancelled","Chargedoff","Current","Defaulted","FinalPaymentInProgress","FinalPaymentInProgress","Past Due (1-15 days)","Past Due (16-30 days)",
                              "Past Due (31-60 days)","Past Due (61-90 days)","Past Due (91-120 days)","Past Due (91-120 days)","Past Due (>120 days)"], axis=1)
df1 = df_af.rename(columns={"Completed": "LoanStatus"})
print(df1)
```

```python
df1.head(15)
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1['BorrowerAPR'] = df1['BorrowerAPR'].fillna(df1['BorrowerAPR'].median())
df1['CreditScoreRangeLower'] = df1['CreditScoreRangeLower'].fillna(df1['CreditScoreRangeLower'].median())
df1['CreditScoreRangeUpper'] = df1['CreditScoreRangeUpper'].fillna(df1['CreditScoreRangeUpper'].median())
df1['DebtToIncomeRatio'] = df1['DebtToIncomeRatio'].fillna(df1['DebtToIncomeRatio'].median())
```

```python
df1['Occupation'] = df1['Occupation'].fillna('Other')
df1['EmploymentStatus'] = df1['EmploymentStatus'].fillna('Not available')
df1['BorrowerState'] = df1['BorrowerState'].fillna('Other')
```
